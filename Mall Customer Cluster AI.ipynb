```python
cust = df[["Annual Income (k$)","Spending Score (1-100)"]].values
```

```python
# cust
```

```python
X = cust
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=5, random_state=42)
```

```python
grup = model.fit_predict(X)
```

```python
model.fit(X)
```

```python
model.predict(X)
```

```python
centro = model.cluster_centers_
```

```python
X[:,0]
```

```python
X[:,1]
```

```python
plt.figure(figsize=(12,8))
plt.scatter(X[:,0], X[:,1], c=grup, s=20)

plt.scatter(centro[:,0], centro[:,1], s=80, c='r', edgecolor='black')
plt.show()
```

```python
df['Cluster'] = grup
```

```python
### Elbow Analysis
```

```python
score = []
for k in range(2, 30):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit_predict(X)
    score.append(kmeans.inertia_) ### Jarak rata" antar data point dalam satu cluster
    
### Semakin kecil Inertia semakin bagus
```

```python
plt.figure(figsize=(12,8))
plt.plot(np.arange(2,30), score, 'ro')
plt.plot(np.arange(2,30), score, 'b-')
plt.xlabel("Jumlah Cluster")
plt.ylabel("Inertia")
plt.show()

## Sebaiknya pilih Jumlah Cluster dimana, penurunan Inertia sudah tidak signifikan
```

```python
from sklearn.metrics import silhouette_score
```

```python
score = []
for k in range(2, 30):
    kmeans = KMeans(n_clusters=k)
    cluster = kmeans.fit_predict(X)
    score.append(silhouette_score(X, cluster))
```

```python
plt.figure(figsize=(12,8))
plt.plot(np.arange(2,30), score, 'ro')
plt.plot(np.arange(2,30), score, 'b-')
plt.xlabel("Jumlah Cluster")
plt.ylabel("Silhouette Score")
plt.show()
### Paling bagus, adalah yg paling tinggi nilai silhouette score
```
